# Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
%matplotlib inline
```

```python
df=pd.read_csv("song_data.csv")
```

```python
df.head()
```

# Preprocessing


```python
sns.histplot(df["prices per unit"],kde=True)
```

```python
df=df[df["prices per unit"]<100]
```

```python
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(),annot=True,cmap="viridis")
```

```python
x=df.drop("prices per unit",axis=1)
y=df["prices per unit"]
```

```python
pca=PCA(.95)
x_pca=pca.fit_transform(x)
```

```python
df["prices per unit"].shape
```

```python
x_pca.shape
```

```python
sns.scatterplot(x=x_pca.reshape(408,),y=df["prices per unit"])
```

# Model Constructor

```python
callback=EarlyStopping(patience=2)
def Function(xtrain,ytrain,xtest,ytest,xval,yval):
    model=Sequential()
    
    model.add(Dense(32,activation="relu"))
    model.add(Dropout(.3))
    
    model.add(Dense(64,activation="relu"))
    model.add(Dropout(.3))
    
    model.add(Dense(32,activation="relu"))
    model.add(Dropout(.3))
    
    model.add(Dense(16,activation="relu"))
    model.add(Dropout(.3))
    
    model.add(Dense(1))
    
    model.compile(loss="mse",metrics=["accuracy"],optimizer="adam")
    
    model.fit(xtrain,ytrain, validation_data=(xval, yval),callbacks=[callback],epochs=600)
    
    return model.history.history
```

**Model 1**

*(only pca)*

```python
xtrain,xtest,ytrain,ytest=train_test_split(x_pca,y,test_size=.2)
```

```python
xtrain,xval,ytrain,yval=train_test_split(xtrain,ytrain,test_size=.1)
```

```python
model1=Function(xtrain,ytrain,xtest,ytest,xval,yval)
```

```python
res1=pd.DataFrame(model1)
```

```python
res1[["accuracy","val_accuracy"]].plot()
```

```python
res1[["loss","val_loss"]].plot()
```
