```python
from PIL import Image
```

```python
for classes in os.listdir("/content/dataset"):
  if classes != ".directory":
    os.mkdir("/content/Resized_img/"+classes)
```

```python
for classes in os.listdir("/content/dataset"):
  if classes != ".directory":
    for img in os.listdir("/content/dataset/"+classes):
      if img != ".directory":
        Image.open("/content/dataset/"+classes+"/"+img).resize((150,150)).save("/content/Resized_img/"+classes+"/"+img)
```

```python
main_dir="/content/Resized_img/"
x=[]
y=[]
for i in os.listdir(main_dir+"0"):
  temp=imread(main_dir+"0/"+i)
  dim1,dim2,channel=temp.shape
  x.append(dim1)
  y.append(dim2)
```

```python
sns.jointplot(x=x,y=y)
```
